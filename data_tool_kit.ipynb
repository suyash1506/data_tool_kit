1. What is NumPy, and why is it widely used in Python?

- NumPy (Numerical Python) is a powerful library for numerical computations in Python. It provides support for arrays, matrices, and many mathematical functions, allowing fast operations on large datasets.

2. How does broadcasting work in NumPy?

- Broadcasting allows NumPy to perform operations on arrays of different shapes. It stretches the smaller array across the larger one without copying data, enabling efficient computation.

3. What is a Pandas DataFrame?

- A Pandas DataFrame is a 2-dimensional, labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table.

4. Explain the use of the groupby() method in Pandas.

- The groupby() method splits the data into groups based on some criteria, applies a function (like mean or sum), and then combines the results. It's useful for summarizing data.

5. Why is Seaborn preferred for statistical visualizations?

- Seaborn provides high-level functions for attractive and informative statistical graphics. It integrates well with Pandas and supports complex visualizations with less code.

6. What are the differences between NumPy arrays and Python lists?

- NumPy arrays are more efficient, support element-wise operations, and require less memory than Python lists. They are ideal for numerical and matrix computations.

7. What is a heatmap, and when should it be used?

- A heatmap is a data visualization that shows the magnitude of a phenomenon as color in two dimensions. It's used for showing correlation matrices or feature intensities.

8. What does the term “vectorized operation” mean in NumPy?

- Vectorized operations are computations applied to entire arrays without writing loops. They are faster and more readable than traditional for-loops.

9. How does Matplotlib differ from Plotly?

- Matplotlib is static and requires more code for customization, while Plotly supports interactive, dynamic plots with built-in zooming and tooltips.

10. What is the significance of hierarchical indexing in Pandas?

- Hierarchical indexing allows multiple index levels on an axis, enabling more complex data structures and easier data manipulation and analysis.
11. What is the role of Seaborn’s pairplot() function?

- pairplot() creates a matrix of scatter plots for all variable combinations in a dataset, useful for exploring relationships between variables.

12. What is the purpose of the describe() function in Pandas?

- describe() generates summary statistics (mean, count, std, min, max, etc.) for numerical columns in a DataFrame.

13. Why is handling missing data important in Pandas?

- Missing data can skew analysis. Pandas offers tools to detect, fill, or drop missing values to maintain data integrity.

14. What are the benefits of using Plotly for data visualization?

- Plotly creates interactive, web-based visualizations with less effort and supports 3D and animated plots, making it powerful for dashboards.

15. How does NumPy handle multidimensional arrays?

- NumPy supports multi-dimensional arrays (ndarrays), allowing operations across any axis, reshaping, slicing, and broadcasting efficiently.

16. What is the role of Bokeh in data visualization?

- Bokeh is a Python library for interactive visualizations in web browsers. It supports streaming and real-time data.

17. Explain the difference between apply() and map() in Pandas.

- map() works on Series for element-wise transformations. apply() can work on Series and DataFrames and is more flexible for applying functions.

18. What are some advanced features of NumPy?

- Advanced features include broadcasting, memory-mapped arrays, linear algebra operations, random sampling, and FFT (Fast Fourier Transform).

19. How does Pandas simplify time series analysis?
- Pandas supports date-time indexing, resampling, rolling statistics, and time zone handling, making time series manipulation easy.

20. What is the role of a pivot table in Pandas?

- Pivot tables summarize data based on groupings and aggregations, helping to restructure and analyze data efficiently.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

- NumPy arrays are stored in contiguous memory blocks and use optimized C operations, making slicing much faster than native Python lists.

22. What are some common use cases for Seaborn?

- Seaborn is used for visualizing distributions, regression plots, categorical data analysis, heatmaps, and complex grid layouts.

23.	How do you create a 2D NumPy array and calculate the sum of each row?

```python
# 23. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np
array = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(array, axis=1)
print(row_sums)
```

```python
# 24. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
mean_value = df['A'].mean()
print(mean_value)
```

```python
# 25.	Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [4, 5, 6]
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
# 26. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'A': [1, 2, 3], 'B': [5, 6, 7]})
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# 27. Generate a bar plot using Plotly.

import plotly.express as px
import pandas as pd
data = {'Fruits': ['Apple', 'Banana', 'Cherry'], 'Count': [10, 15, 7]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Fruits', y='Count')
fig.show()
```

```python
# 29. Write a program to perform element-wise multiplication of two NumPy arrays .

import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
# 30.	Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [2, 5, 10]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
# 31. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30]})
filtered_df = df[df['A'] > 15]
print(filtered_df)
```

```python
# 32. Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
sns.histplot(data, bins=5, kde=True)
plt.show()
```

```python
# 33.	Perform matrix multiplication using NumPy.

import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```
